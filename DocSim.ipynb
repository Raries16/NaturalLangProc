```python
def cleaned_text(text):
    txt_1 = text.lower()
    txt_2 = re.sub(r'[^a-zA-Z0-9]'," ",txt_1)
    words = re.compile('effectiveuncontrolled copy')
    txt_3 = re.sub(words, '', txt_2)
    txt_4= re.sub(r'\bpage            of    \d\d\d',"",txt_3)
    txt_5= re.sub(r'effective from  \d\d \w\w\w \d\d\d\d',"",txt_4)
    txt_6 = re.sub(r'doc  no   sop',"", txt_5)
    txt_7 = re.sub(r'version',"", txt_6)
    txt_8 = re.sub(r'last periodic review date',"",txt_7)
    txt_9 = re.sub(r'approvals',"",txt_8)
    txt_10 = re.sub(r'name  intentions  signatures  date  utc',"",txt_9)
    return(txt_10)
```

# Soft Cosine Using Fasttext

```python
import gensim.downloader as api
from gensim import corpora
from gensim.matutils import softcossim
```

```python
# Download the FastText model
fasttext_model300 = api.load('fasttext-wiki-news-subwords-300')
```

```python
# Prepare a dictionary and a corpus.
documents = [text_12 , text_22]
dictionary = corpora.Dictionary(documents)
```

```python
# Prepare the similarity matrix
similarity_matrix = fasttext_model300.similarity_matrix(dictionary)
```

```python
# Convert the sentences into bag-of-words vectors.
sent_1 = dictionary.doc2bow(text_12)
sent_2 = dictionary.doc2bow(text_22)
```

```python
# Compute soft cosine similarity
print(softcossim(sent_1, sent_2, similarity_matrix))
```

# Gensim LSI


```python
from gensim import corpora, models, similarities
from collections import defaultdict
```

```python
document1 = [text_13 , text_23]
```

```python
frequency = defaultdict(int)
for text in document1:
    for token in text:
        frequency[token] += 1
document1 = [[token for token in text if frequency[token] > 0]
         for text in document1]
dictionary = corpora.Dictionary(document1)
```

```python
corpus = [dictionary.doc2bow(text) for text in document1]
lsi = models.LsiModel(corpus, id2word=dictionary)
doc = text_13
vec_bow = dictionary.doc2bow(doc.lower().split())
```

```python
vec_lsi = lsi[vec_bow]
index = similarities.MatrixSimilarity(lsi[corpus])
```

```python
sims = index[vec_lsi]
sims = sorted(enumerate(sims), key=lambda item: -item[1])

print(sims)
```
